```python
def plot_cf(train):
    fig = plt.figure(figsize=(20,8))
    
    ax1 = fig.add_subplot(211)
    fig = sm.graphics.tsa.plot_acf(train, lags=20,ax=ax1)
    ax1.xaxis.set_ticks_position('bottom')
    fig.tight_layout()
    
    ax2 = fig.add_subplot(212)
    fig = sm.graphics.tsa.plot_pacf(train, lags=20, ax=ax2)
    ax2.xaxis.set_ticks_position('bottom')
    fig.tight_layout()
    plt.show()
plot_cf(train)
```

```python
start = datetime.datetime(2016, 1, 1) # or start = '1/1/2016'
end = datetime.date.today()
prices = web.DataReader('AAPL', 'yahoo', start, end)

# 
train_p = prices[:-150]['Close'].to_list()
test_p = prices[-150:]['Close'].to_list()

# 
model = ARIMA(train_p, order=(20,2,20))
fitted = model.fit()
pred = list(fitted.forecast(150))

# 
plt.figure(figsize=(10,6))
plt.grid(True)
plt.xlabel('Dates')
plt.ylabel('Closing Prices')
plt.plot(train_p + test_p, 'green', label='Train data')
plt.plot(train_p + pred, 'red', label='pred')
plt.plot(train_p , 'blue', label='Test data')
plt.legend()
```

### EEMD

```python
from PyEMD import EMD
from PyEMD import EMD, Visualisation
from PyEMD import CEEMDAN
```

```python
emd = EMD()
IMFs = emd(np.array(train_p))

fig = plt.figure('EEMD', figsize=(10, 20))
for i in range(9):
    ax = fig.add_subplot(10,1,i+1)
    ax.plot(IMFs[i])
plt.title('EEMD')
plt.show()
```

```python
ceemdan = CEEMDAN()
cIMFs = ceemdan(np.array(train_p))

fig = plt.figure('cEEMD', figsize=(10, 20))
for i in range(9):
    ax = fig.add_subplot(10,1,i+1)
    ax.plot(cIMFs[i])
plt.title('cEEMD')
plt.show()
```

```python
t = np.arange(0, len(train_p), 1)

emd1 = EMD()
emd1.emd(np.array(train_p))
imfs, res = emd1.get_imfs_and_residue()

plt.figure(figsize=(20, 10))
vis = Visualisation()
vis.plot_imfs(imfs=imfs, residue=res, t=t, include_residue=True)
vis.plot_instant_freq(t, imfs=imfs)
vis.show()
```
